```python
import numpy as np
import keras
from keras.utils import np_utils
from keras.optimizers import Adam
import cv2
```

```python
def load_img(f):
    f=open(f)
    lines=f.readlines()
    
    imgs,lab=[], []
    
    for i in range(len(lines)):
        fn,label = lines[i].split(' ')
        im1 = cv2.imread("Allandold/"+fn)
        im1 = cv2.resize(im1,(128,128))
        print (i,'Allandold/'+fn)
        imgs.append(im1)
        lab.append(int(label))
        
    return imgs, lab
```

```python
train_x,train_y=load_img('Allandold/train.txt')
```

```python
test_x,test_y=load_img('Allandold/test_processed.txt')
```

```python
print ('have %d train img and %d test img.' % (len(train_x),len(test_x)))
```

```python
train_x = np.array(train_x)
train_y = np.array(train_y)
test_x = np.array(test_x)
test_y = np.array(test_y)
```

```python
print (test_y.shape)
```

```python
#Process data
train_x = train_x.reshape(-1, 128, 128, 3) # Expend dimension for 1 cahnnel image
test_x = test_x.reshape(-1, 128, 128, 3)  # Expend dimension for 1 cahnnel image
train_x = train_x / 255 # Normalize
test_x = test_x / 255 # Normalize
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(train_x, train_y, test_size=0.20, random_state=42)
```

```python
print (X_train.shape[1:])
```

```python
#One hot encoding
y_train = np_utils.to_categorical(y_train, num_classes=2)
y_test = np_utils.to_categorical(y_test, num_classes=2)

```

```python
print (X_train.shape)
```

model

```python
from keras.layers import Dense, Activation, Convolution2D, MaxPooling2D, Flatten, Dropout
from keras.optimizers import adam
from keras.models import Sequential
```

```python
model = Sequential()  
# Create CN layer 1  
model.add(Convolution2D(filters=16,  
                 kernel_size=(5,5),  
                 padding='same',  
                 input_shape=(128,128,3),  
                 activation='relu'))  # (16,124,124)
# Create Max-Pool 1  
model.add(MaxPooling2D(pool_size=(2,2)))  # (16,62,62)
  
# Create CN layer 2  
model.add(Convolution2D(filters=36,  
                 kernel_size=(5,5),  
                 padding='same',  
                 input_shape=(128,128,3),  
                 activation='relu'))  # (36,58,58)
  
# Create Max-Pool 2  
model.add(MaxPooling2D(pool_size=(2,2)))  # (36,29,29)
  
# Add Dropout layer  
model.add(Dropout(0.25))  
```

```python
model.add(Flatten())
model.add(Dense(8196, activation='relu'))  
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))
model.summary()
```

```python
from keras import optimizers

sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
train_history = model.fit(x=X_train,  
                          y=y_train, validation_data=(X_test, y_test),  
                          epochs=10, batch_size=32, verbose=2)  
```

```python
model.save("i-CNN-model.h5")
```

```python
#Plot Loss and accuracy
import matplotlib.pyplot as plt
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.plot(train_history.history['accuracy'])
plt.plot(train_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')

plt.subplot(1,2,2)
plt.plot(train_history.history['loss'])
plt.plot(train_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
plt.show()
```

```python
test_y = list(map(int,test_y))
test_y = np_utils.to_categorical(test_y)
```

```python
eval = model.evaluate(test_x, test_y)

print( 'loss =', eval[0] )
print( 'accuracy =', eval[1] ) 
```
